# Practicing with Pandas

Lets first import pandas

`````python
import pandas as pd
`````

Now we can read out excel file

`````python
array1 = pd.read_excel("../Excel_Sheets/Fruit-Prices-2022.xlsx")
`````

`````python
array1 # Lets view our aray
`````

As you can see we have some unneccessary columns at the end so we will delete them. First lets see what they are called using the ```info``` funciton. While we are doing this we can see the data types of all the columns.

`````python
array1.info()
`````

`````python
array2 = array1.drop(['Unnamed: 8', 'Unnamed: 9',"Unnamed: 10"], axis=1)
`````

`````python
array2
`````

To generate some summary statistics we will use the ```describe``` function

`````python
array2["RetailPrice"].describe()
`````

`````python
array2["CupEquivalentPrice"].describe()
`````

We can locate specific values using ```loc``` and ```iloc```

`````python
array2.loc[2, "Fruit"]
`````

`````python
array2.iloc[2, 1]
`````

To modify the data we can use a for loop and append, and then add them to our array.

`````python
n = 0
tasterating = []
for n in range(62):
  n += 1
  tasterating.append(n)
  
array2['TasteRating'] = tasterating
`````

Lets look at our new column

`````python
array2.info()
`````

To check if we have any Nulls or NaNs in our code we will use the ```isnull``` function.

`````python
array2.isnull()
`````

To make sure we do not have any duplicates in our table we will use the function ````drop_duplicates``` to remove them all.

`````python
array2.drop_duplicates()
`````

Now lets make some graphs with our data. We will be using seaborn for this.

`````python
import seaborn as sns
import matplotlib.pyplot as plt
`````

This is our data set.

`````python
array2
`````

`````python
sns.catplot(data=array2, kind="bar", x="Form", y="CupEquivalentPrice", hue="RetailPrice")
`````
